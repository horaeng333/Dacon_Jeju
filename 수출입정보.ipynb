````python
df_test['품목명'].value_counts()
````

````python
df_test = df_test[df_test['품목명']!='무화과']
df_test
````

````python
df_test.loc[df_test['품목명']=='꽃양배추와 브로콜리(broccoli)','품목명'] = '브로콜리'
df_test
````

````python
df_test.loc[df_test['품목명']=='방울다다기 양배추','품목명'] = '양배추'
df_test
````

````python
df_test.loc[df_test['품목명']=='순무','품목명'] = '무'
df_test
````
